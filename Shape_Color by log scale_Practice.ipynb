To complete the second task, we need to first reshape the dataset so that all Pokémon types are recorded in a single column. This will add duplicates of Pokémon with two types, which is fine for the task to be performed.

```python
pokemon.head()
```

```python
pokemon.columns
```

```python
type_cols = ['type_1','type_2']
```

```python
non_type_cols = pokemon.columns.difference(type_cols)  # To combine type 1 and type 2, set difference()
non_type_cols
```

```python
pkmn_types = pokemon.melt(id_vars = non_type_cols, value_vars = type_cols, 
                          var_name = 'type_level', value_name = 'type').dropna()
pkmn_types.head()
```

### By Shape

```python
ttype_markers = [['fairy', 'o'], ['dragon', '^']]

for ttype, marker in ttype_markers:
    plot_data = pkmn_types.loc[pkmn_types['type'] == ttype]   # plot for (1) fairy (2) dragon as 'o' or '^'
    sb.regplot(data = plot_data,
               x = 'weight', 
               y = 'height',
               x_jitter = 0.04,
               fit_reg = False, 
               marker = marker)
    
plt.legend(['fairy', 'dragon'])
```

### By Color

```python
subset = pkmn_types.loc[pkmn_types['type'].isin(['fairy','dragon'])]
subset.type.value_counts()
```

```python
g = sb.FacetGrid(data = subset, hue = 'type', 
                 hue_order = ['fairy', 'dragon'], size = 5, aspect= 1.5);

g.map(sb.regplot, 'weight', 'height', x_jitter = 0.04, fit_reg = False)

x_ticks = [0.1, 0.3, 1, 3, 10, 30, 100, 300, 1000]
g.set(xscale = 'log')   # change scale to log
g.set(xticks = x_ticks, xticklabels = x_ticks)

g.add_legend() 
```
