```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
pokemon = pd.read_csv('pokemon.csv')
```

```python
bins = np.arange(0, pokemon['height'].max()+0.5, 0.5)
plt.hist(data = pokemon, x = 'height', bins = bins);
```

```python
bins = np.arange(0, pokemon['height'].max()+0.2, 0.2)
plt.hist(data = pokemon, x = 'height', bins = bins);
plt.xlim((0,6)); # 특정 범위만 표기
```

```python
plt.figure(figsize = [20,5])

plt.subplot(1,2,1)
bins = np.arange(0, pokemon['height'].max()+0.5, 0.5)
plt.hist(data = pokemon, x = 'height', bins = bins);

plt.subplot(1,2,2)
bins = np.arange(0, pokemon['height'].max()+0.2, 0.2)
plt.hist(data = pokemon, x = 'height', bins = bins);
plt.xlim((0,6)); # 특정 범위만 표기
```

```python
bins = np.arange(0, pokemon['weight'].max()+40, 40)
plt.hist(data = pokemon, x = 'weight', bins = bins);
```

```python
bins = np.arange(0, pokemon['weight'].max()+40, 40)
plt.hist(data = pokemon, x = 'weight', bins = bins);
plt.xscale('log'); # 로그 스케일로 변경
```

```python
np.log10(pokemon.weight.describe())
```

```python
bins = 10 ** np.arange(-1, 3 + 0.1, 0.1)
plt.hist(data = pokemon, x = 'weight', bins = bins);
plt.xscale('log'); # 로그 스케일로 변경. Scale down!
```

```python
bins = 10 ** np.arange(-1, 3 + 0.1, 0.1)   # revert to integer

ticks = [0.1, 0.3, 1, 3, 10, 30, 100, 300, 1000]
labels =['{}'.format(v) for v in ticks]

plt.hist(data = pokemon, x = 'weight', bins = bins);
plt.xscale('log'); # 로그 스케일로 변경

plt.xticks(ticks, labels); # 차트 다음에 위치시켜야 출력됨
#For the ticks, we have used xticks() to specify locations and labels in their natural units. 
```

After applying a logarithmic transform to the data, the data will follow a **normal distribution**
